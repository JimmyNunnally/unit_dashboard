```python
#@title Russian personnel losses 
df = person.copy()
fig = go.Figure(data=[go.Bar(x=df['date'],y=df['personnel'],name='personnel',marker={'color': 'red'}),
                          go.Bar(x=df['date'],y=df['POW'],name='POW',marker={'color':' orange'})])
                      
fig.add_trace(go.Scatter(x=df['date'],y=df['personnel'],mode='lines+markers',
                    name='personnel',marker={'color': 'black'}))
fig.add_trace(go.Scatter(x=df['date'],y=df['POW'],mode='lines+markers',
                    name='POW',marker={'color': 'black'}))
fig.update_layout(
    autosize=False,
    title ="Russian loss of Humans",title_x=0.5,
    width=900,
    height=550,
xaxis_title="Date",
yaxis_title="Number of People ", paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)')
fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="personnel",
                     method="update",
                     args=[{"visible": [True,False]},
                           {"title": "Human Loss"}]),
                dict(label="POW",
                     method="update",
                     args=[{"visible": [False,True]},
                           {"title": "Prisoners of WAR",
                            }]),
            ]),
        type="dropdown",
        showactive=True,
        x=0.05,
        xanchor="left",
        y=1,
        yanchor="top",bgcolor='lightblue' )
    ])
```

```python
import pandas as pd
uk=pd.read_csv('/content/losses_ukraine.csv')
```

```python
tanks = uk[uk["equipment"] == "Tanks"]

fig = px.bar(tanks, x="model", y="losses_total", color="model", title = "Ukraine Tanks Total Losses by Model")
fig.show()
```

```python
import matplotlib.pyplot as plt
t64bv = uk[uk["model"] == "T-64BV"]

t64bv.plot(x = "model", y = ["destroyed", "captured", "abandoned"], kind = "bar", 
           color = ["#C00000", "#0000A1", "#006900"])

plt.title("Ukraine T-64BV Tank Total Losses")
plt.grid(axis = "y")
plt.ylabel("Count")
plt.show()
```
